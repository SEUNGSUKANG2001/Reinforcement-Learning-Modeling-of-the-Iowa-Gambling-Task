```python
import json
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
with open('learners_sub.json', 'r', encoding='utf-8') as f:
    parameters = json.load(f)
print(parameters[0]) ## 잘 되는 지 예시
```

```python
df = pd.DataFrame(parameters)
df.head()
```

```python
plot_standard = ['q_alpha', 'q_beta', 'vs_alpha_plus', 'vs_alpha_minus', 'vs_beta', 'vs_w', 'nll_q', 'nll_vs']
#각 값 별로 분포 확인
for ps in plot_standard:
    plt.title(f"log_{ps}")
    plt.hist(np.log10(df[ps]+1e-12), bins = 30)
    plt.show()
    plt.title(f"{ps}")
    plt.hist(df[ps], bins = 15)
    plt.show()
```
