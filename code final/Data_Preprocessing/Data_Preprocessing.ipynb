```python
#데이터를 Preprocessing하여 상위 폴더에 저장 

import pandas as pd 

#100개짜리 choice dict
df_choice_100 = pd.read_csv("choice_100.csv")
choice_100_dict = {
    i + 1: list(row) for i, (idx, row) in enumerate(df_choice_100.iterrows())
}

#100개짜리 gain dict
df_gain_100 = pd.read_csv("wi_100.csv")
gain_100_dict = {
    i + 1: list(row) for i, (idx, row) in enumerate(df_gain_100.iterrows())
}

#100개짜리 loss dict
df_loss_100 = pd.read_csv("lo_100.csv")
loss_100_dict = {
    i + 1: list(row) for i, (idx, row) in enumerate(df_loss_100.iterrows())
}

#최종 dictionary : 후에 150짜리 이어붙임
'''key : [
  [1, 2, 3, ...],          # choice
  [100, 100, 100, ...],    # gain
  [-250, 0, -50, ...],     # loss
  [350, 100, 150, ...]     # gain + loss
]
'''
final_dict = {
    i: [choice_100_dict[i],
        [g for g in gain_100_dict[i]],
        [l for l in loss_100_dict[i]],
        [(g + l) for g, l in zip(gain_100_dict[i], loss_100_dict[i])]]
    for i in choice_100_dict
}
```

```python
#150개짜리 choice dict
df_choice_150 = pd.read_csv("choice_150.csv")
df_choice_150to100 = df_choice_150.loc[:, 'Choice_1':'Choice_100'] #100까지만
choice_150_dict = {
    i + 1: list(row) for i, (idx, row) in enumerate(df_choice_150to100.iterrows())
}

#150개짜리 gain dict
df_gain_150 = pd.read_csv("wi_150.csv")
df_gain_150to100  = df_gain_150.loc[:, 'Wins_1':'Wins_100'] #100까지만
gain_150_dict = {
    i + 1: list(row) for i, (idx, row) in enumerate(df_gain_150to100.iterrows())
}
#150개짜리 loss dict
df_loss_150 = pd.read_csv("lo_150.csv")
df_loss_150to100  = df_loss_150.loc[:, 'Losses_1':'Losses_100'] #100까지만
loss_150_dict = {
    i + 1: list(row) for i, (idx, row) in enumerate(df_loss_150to100.iterrows())
}

final_dict_150 = {
    i: [choice_150_dict[i],
        [g for g in gain_150_dict[i]],
        [l for l in loss_100_dict[i]],
        [(g + l) for g, l in zip(gain_150_dict[i], loss_150_dict[i])]]
    for i in choice_150_dict
}
```

```python
#후에 150 이어붙임
final_dict.update({len(final_dict) + i + 1: v for i, (_, v) in enumerate(final_dict_150.items())})
```

```python
import json
import os

# 현재 경로 기준 상위 폴더 경로 구하기
parent_dir = os.path.abspath(os.path.join(os.getcwd(), os.pardir))

# 상위 폴더 내 "IGT_Parameter Estimation" 폴더 경로
target_dir = os.path.join(parent_dir, "IGT_Parameter Estimation_and_clustering")

# 폴더가 없으면 생성
os.makedirs(target_dir, exist_ok=True)

# 저장할 파일 경로 설정
file_path = os.path.join(target_dir, "data.json")

# JSON 파일로 저장
with open(file_path, "w", encoding="utf-8") as f:
    json.dump(final_dict, f, indent=4, ensure_ascii=False)
```
